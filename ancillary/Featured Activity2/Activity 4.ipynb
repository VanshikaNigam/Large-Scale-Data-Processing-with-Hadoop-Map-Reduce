<b>Activity 4a) Bigram </b>

```r
Time<-c("1.21","1.51","2.10","2.12","9.14")
```

```r
Num_Files<-c(2,4,6,7,15)
Num_Files
```

```r
Time_plot<-cbind(Num_Files,Time)
Time_plot<-as.data.frame(Time_plot)
Time_plot
```

<b>The Time increases as the number of file sto be processed by MR increases.</b>

```r
plot(x = Num_Files,y = Time,type="b" ,col="green",xlab = "Number of Files" ,ylab = "Time Taken in minutes")
```

<b>Activity 4b) Trigram</b>

```r
Time<-c("3.11","4.09","4.89","5.12","12.55")
Num_Files<-c(2,4,6,7,15)
```

```r
Time_plot<-cbind(Num_Files,Time)
Time_plot<-as.data.frame(Time_plot)
Time_plot
```

```r
plot(x = Num_Files,y = Time,type="b" ,col="blue",xlab = "Number of Files" ,ylab = "Time Taken")
```
